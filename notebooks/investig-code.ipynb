```python
import torch
import cv2
import os
import torchvision.utils as vutils
from matplotlib import pyplot as plt
import numpy as np
```

```python
mask_fn = os.listdir("data/processed/mask")
```

```python
for i, fn in enumerate(mask_fn):
    mfp = os.path.join("data/processed/mask", fn)
    imgfp = os.path.join("data/processed/img", fn)
    img = cv2.imread(imgfp, cv2.IMREAD_GRAYSCALE)
    mask = cv2.imread(mfp, cv2.IMREAD_GRAYSCALE)
    
    print(i)
    plt.subplot(121)
    plt.imshow(img)
    plt.subplot(122)
    plt.imshow(mask)
    plt.show()
```
